```python
import pandas as pd

df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_259124.csv", skiprows=4)  # Adjust the number of rows to skip
print(df.head())
```

```python
year = '2016'
```

```python
population_data = df[['Country Name', year]].dropna()
```

```python
population_data.sort_values(by=year, ascending=False, inplace=True)
```

```python
N = 10
top_countries = population_data.head(N)
```

```python
plt.figure(figsize=(12, 8))
plt.bar(top_countries['Country Name'], top_countries[year], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title(f'Top {N} Countries by Population in {year}')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
